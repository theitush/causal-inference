### Naive - bootstrap CI estimation

```python
def CI_95_estimation(mean, variance):
    return {'mean': mean, 'lower': mean - 1.96*variance**0.5, 'upper': mean + 1.96*variance**0.5}
```

```python
def calc_naive_ATE(data):
    naive_death_rates = data.groupby('treatment').death.mean()
    naive_ATE = naive_death_rates[1]-naive_death_rates[0]
    return naive_ATE
```

```python
def bs_CI_naive_estimation(data, repetitions=500):
    bs_ates = []
    for i in range(repetitions):
        sample = data.sample(n=len(data), replace=True)
        bs_ates.append(calc_naive_ATE(sample))
    return CI_95_estimation(np.mean(bs_ates), np.var(bs_ates))
```

```python
bs_CI_naive_estimation(df)
```

### Naive - asymptotic CI estimation

```python
treated = df[df.treatment==1]
untreated = df[df.treatment==0]
naive_variance = (treated.death.var() / len(treated)) + (untreated.death.var() / len(untreated))
naive_variance
```

```python
CI_95_estimation(calc_naive_ATE(df),naive_variance) 
```

### Standardization - bootstrap CI estimation

```python
def calc_standard_ATE(data):
    sepsis_std_death_rate = data.groupby(['treatment','sepsis']).death.mean()
    sepsis_std_ATE = sepsis_std_death_rate[1].sum() - sepsis_std_death_rate[0].sum()
    return sepsis_std_ATE


def bs_CI_std_estimation(data, repetitions=500):
    bs_ates = []
    for i in range(repetitions):
        sample = data.sample(n=len(data), replace=True)
        bs_ates.append(calc_standard_ATE(sample))
    return CI_95_estimation(np.mean(bs_ates), np.var(bs_ates))
```

```python
bs_CI_std_estimation(df)
```

### Standardization - asymptotic CI estimation

```python
def estimate_var(data):
    return data.death.var() / len(data)
```

```python
sepsis_freq = df.sepsis.mean()
sepsis_ATE = calc_standard_ATE(df)
variances = df.groupby(['treatment','sepsis']).apply(estimate_var)
variances
```

```python
# from Q5 we know:
stdr = df.groupby(['treatment','sepsis']).death.mean()
sepsis_variance = ((stdr[1][1]-stdr[1][0]+stdr[0][0]-stdr[0][1])**2) * (sepsis_freq * (1-sepsis_freq) / len(df)) + (
    ((1-sepsis_freq)**2) * (variances[1][0] + variances[0][0])) + (
    (sepsis_freq**2) * (variances[1][1] + variances[0][1]))
```

```python
CI_95_estimation(sepsis_ATE, sepsis_variance)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df['treatment_X_female'] = df['treatment'] * df['female']
df['age^2'] = df['age'] ** 2
treatment = df[df.treatment==1]
ctrl = df[df.treatment==0]
```

```python
X = ['treatment_X_female', 'ARF','age','age^2','sepsis']
y = ['death']
```

```python
lr_treated = LogisticRegression(random_state=0).fit(treatment[X], treatment[y].values.ravel())
lr_ctrl = LogisticRegression(random_state=0).fit(ctrl[X], ctrl[y].values.ravel())
```

```python
df['treated_predict'] = lr_treated.predict(df[X])
df['ctrl_predict'] = lr_ctrl.predict(df[X])
```

```python
mean_treated = df['treated_predict'].mean()
mean_ctrl = df['ctrl_predict'].mean()
```

```python
mean_treated - mean_ctrl  
```

```python
mean_treated / mean_ctrl
```

```python
(mean_treated/(1-mean_treated)) / (mean_ctrl/(1-mean_ctrl))
```

```python
def lr_bootstrap_CI_estimation(data, repetitions=300):
    X = ['treatment_X_female', 'ARF','age','age^2','sepsis']
    y = ['death']
    diff_ates = []
    ratio_ates = []
    odds_ratio_ates = []
    for i in range(repetitions):
        sample = data.sample(n=len(data), replace=True)
        treatment = sample[sample.treatment==1]
        ctrl = sample[sample.treatment==0]
        lr_treated = LogisticRegression(max_iter=500).fit(treatment[X], treatment[y].values.ravel())
        lr_ctrl = LogisticRegression(max_iter=500).fit(ctrl[X], ctrl[y].values.ravel())
        sample['treated_predict'] = lr_treated.predict(sample[X])
        sample['ctrl_predict'] = lr_ctrl.predict(sample[X])
        mean_treated = sample['treated_predict'].mean()
        mean_ctrl = sample['ctrl_predict'].mean()
        diff_ates.append(mean_treated - mean_ctrl)
        ratio_ates.append(mean_treated / mean_ctrl)
        odds_ratio_ates.append((mean_treated/(1-mean_treated)) / (mean_ctrl/(1-mean_ctrl)))
    return {'diff_ate': CI_95_estimation(np.mean(diff_ates), np.var(diff_ates)),
            'ratio_ates': CI_95_estimation(np.mean(ratio_ates), np.var(ratio_ates)),
            'odds_ratio_ates': CI_95_estimation(np.mean(odds_ratio_ates), np.var(odds_ratio_ates))}
```

```python
lr_bootstrap_CI_estimation(df)
```

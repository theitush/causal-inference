```python
import pandas as pd
```

```python
data = pd.DataFrame()
data['X'] = [0,0,0,0,1,1,1,1]
data['A'] = [0,0,1,1,0,0,1,1]
data['Y'] = [0,1,0,1,0,1,0,1]
amounts = [82,24,193,93,172,263,49,124]
data
```

```python
df = pd.DataFrame()
for i, amount in enumerate(amounts):
    df = pd.concat([df, pd.concat([data.loc[i]]*amount, axis=1).T])
df = df.reset_index()
```

```python
df.groupby(['X','A','Y']).count()
```

```python
df.groupby(['A'])['Y'].mean()
```

```python
df.groupby(['X','A'])['Y'].mean()
```

```python
def calc_ATE(df):
    return df[df['A']==1]['Y'].mean() - df[df['A']==0]['Y'].mean()
```

```python
ATE = calc_ATE(df)
CATE = df.groupby('X').apply(calc_ATE)
len_Xs = df.groupby('X').Y.count()
ATE_conditional = ((CATE[0] * len_Xs[0]) + (CATE[1] * len_Xs[1])) / sum(len_Xs)
```

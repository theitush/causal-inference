```python
# import R's utility package
utils = rpackages.importr('utils')
# select a mirror for R packages
utils.chooseCRANmirror(ind=1) # select the first mirror in the list
```

```python
from contextlib import contextmanager
from rpy2.robjects.lib import grdevices
from IPython.display import Image, display

@contextmanager
def r_inline_plot(width=600, height=600, dpi=100):

    with grdevices.render_to_bytesio(grdevices.png, 
                                     width=width,
                                     height=height, 
                                     res=dpi) as b:

        yield

    data = b.getvalue()
    display(Image(data=data, format='png', embed=True))
```

## Loading the Dataset

```python
robjects.r('''rhc.data <- read.csv2("../rhc.csv") %>% filter(meanbp1 > 0)    
    rhc.data<-na.omit(rhc.data)
    rhc.data$ID <- 1:nrow(rhc.data)
    rhc.data$Gender <- factor(rhc.data$female, levels =  c(0,1), labels = c("Male", "Female"))    
    rhc.data$Treatment <- factor(na.omit(rhc.data$treatment), levels =  c(0,1), labels = c("Untreated", "Treated"))
    set.seed(31415)
''')
```

```python
robjects.r('''str(rhc.data)
    summary(rhc.data)''')
```

## Exact Matching

```python
treatment = np.array(robjects.r('''rhc.data$treatment'''))
treatment[1942]
# len(treatment)
np.unique(treatment)
# np.where(np.logical_and(treatment != 1, treatment!=2))
```

```python
a = robjects.r('''rhc.data$Treatment <- na.omit(rhc.data$Treatment)''')
# Counter(robjects.r('''rhc.data$Treatment'''))
Counter(a)
```

```python
robjects.r('''m.exact <- matchit(formula = Treatment ~ age + meanbp1 + female, method = "exact", data = rhc.data)
              m.exact
              rhc.data <- rhc.data %>% mutate(age.rnd = round(age,0)) # Add a rounded version of age
              m.exact <- matchit(formula = treatment ~ age.rnd + meanbp1 + female, method = "exact", data = rhc.data)
              m.exact # Better, but still not so useful  
              summary(m.exact)              
              #plot(m.exact)
              matched.rhc.exact <- match.data(m.exact)
              matched.rhc.exact %>% select(subclass, age.rnd, meanbp1, female, treatment, death) %>% head()
              matched.rhc.exact %>% filter(subclass==363)''')
```

```python
with r_inline_plot(width=900, height=600, dpi=50):
    robjects.r('''rhc.data$type <- "Original Data"
    matched.rhc.exact$type = "Exactly Matched Data"
    df.for.plot <- bind_rows(rhc.data, matched.rhc.exact) 
    p <- ggplot(df.for.plot, aes(x = age, y = meanbp1, color = Gender, 
                                                          shape = Treatment)) +
      theme_bw() +  geom_point(size = 3) + facet_wrap(~type) + 
      theme(title =  element_text(size=22), strip.text = element_text(size =20), 
            legend.text = element_text(size = 20),
            legend.position="bottom") + 
      ylab("Mean Blood Pressure Day 1") + xlab("Age")  
      p <- p + geom_point()
      print(p)''')
```

## Nearest neighbor Matching ##

```python
robjects.r('''m.nn <- matchit(formula = Treatment ~ age + meanbp1 + female, method = "nearest", 
                     distance = "mahalanobis", data = rhc.data)
    m.nn %>% summary
    m.nn %>% summary(standardize = TRUE)
    # Trick to get the matched sets variable, which match.data() does not report for nn
    temp.df <- as.data.frame(m.nn$match.matrix)
    colnames(temp.df) <- c("matched.unit")
    temp.df$untrt.unit <- as.numeric(as.character(temp.df$matched.unit))
    temp.df$trt.unit <- as.numeric(rownames(temp.df))
    #now delete matches=na
    temp.mathces.only <- temp.df %>% filter(!is.na(matched.unit))
    temp.mathces.only$subclass.nn <- 1:nrow(temp.mathces.only)
    head(temp.mathces.only)
    rhc.data$subclass.nn <- 0
    rhc.data$subclass.nn[temp.df$trt.unit]<- temp.mathces.only$subclass.nn
    rhc.data$subclass.nn[temp.df$untrt.unit]<- temp.mathces.only$subclass.nn
    # Keep matched data only
    matched.rhc.nn <- rhc.data %>% filter(subclass.nn>0)''')
```

```python
with r_inline_plot(width=900, height=600, dpi=50):
    robjects.r('''matched.rhc.nn$type <- "NN Matched Data"        
        df.for.plot <- bind_rows(rhc.data, df.for.plot)
        p <- ggplot(df.for.plot, aes(x = age, y = meanbp1, color = Gender, 
                                shape = Treatment)) +
          theme_bw() +  geom_point(size = 3) + facet_wrap(~type) + 
          theme(title =  element_text(size=22), strip.text = element_text(size =20), 
                legend.text = element_text(size = 20),
                legend.position="bottom") + 
          ylab("Mean Blood Pressure Day 1") + xlab("Age")

          matched.rhc.nn %>% filter(subclass.nn>0) %>%  ggplot(aes(x = age, y = meanbp1, color = Gender, 
                                shape = Treatment)) +
          theme_bw() +  geom_point(size = 3) +   geom_line(aes(group=subclass.nn)) + 
          theme(title =  element_text(size=22), strip.text = element_text(size =20), 
                legend.text = element_text(size = 20),
                legend.position="bottom") + 
          ylab("Mean Blood Pressure Day 1") + xlab("Age")
          p <- p + geom_point()
          print(p)''')
```

## Optimal Matching - Mahalanobis Distance

```python
with r_inline_plot(width=900, height=600, dpi=50):
    robjects.r('''m.opt <- matchit(formula = treatment ~ age + meanbp1 + female, method = "optimal", 
                            distance = "mahalanobis", data = rhc.data)

            matched.rhc.opt <- match.data(m.opt)

            matched.rhc.opt %>% filter(subclass==1)
            matched.rhc.opt %>% filter(subclass==2)

            m.opt %>% summary
            m.opt %>% summary(standardize = TRUE)

            matched.rhc.opt$type <- "Optimal Matched Data"
            df.for.plot <- bind_rows(df.for.plot, matched.rhc.opt)
            p <- ggplot(df.for.plot, aes(x = age, y = meanbp1, color = Gender, 
                                    shape = Treatment)) + 
              theme_bw() +  geom_point(size = 3) + facet_wrap(~type) + 
              theme(title =  element_text(size=22), strip.text = element_text(size =20), 
                    legend.text = element_text(size = 20),
                    legend.position="bottom") + 
              ylab("Mean Blood Pressure Day 1") + xlab("Age")
            p <- p + geom_point()
            print(p)''')
```

```python
robjects.r('''

    # Simple estimators in the matched dataset:
    mean(matched.rhc.nn$death[matched.rhc.nn$treatment==1] - matched.rhc.nn$death[matched.rhc.nn$treatment==0])
    t.test(matched.rhc.nn$death[matched.rhc.nn$treatment==1], matched.rhc.nn$death[matched.rhc.nn$treatment==0], paired = T)
    wilcox.test(matched.rhc.nn$death[matched.rhc.nn$treatment==1], matched.rhc.nn$death[matched.rhc.nn$treatment==0], paired = T)
    ''')
```
